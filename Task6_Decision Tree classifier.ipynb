### Task 6 : Create the Decision Tree classifier and visualize it graphically.

#### Objective:The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
# importing labriries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Iris.csv')# loading the file using read_csv command
```

```python
iris.head()# will give you the first 5 rows of the dataset
```

```python
iris.tail()# will give you the last 5 rows of the dataset 
```

```python
iris.isnull().sum() # this command will show total number of null values in particular column
```

```python
iris.Species.value_counts()# this will show total count of each species
```

```python
iris.info()
```

```python
iris.describe()
```

```python
x = iris.iloc[:,[0,1,2,3]].values
y = iris.Species.values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier(criterion ='entropy')
dtclassifier.fit(x_train,y_train)
```

```python
y_pred = dtclassifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
result = accuracy_score(y_test,y_pred)
print('Accuracy Score is : ',result*100)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn import tree
```

```python
features = ['sepal length','sepal lwidth','petal length','petal width']
species = ['Iris-versicolor','Iris-setosa','Iris-virginica']

fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(15,10),dpi=350)
tree.plot_tree(dtclassifier,feature_names = features,class_names = species,filled = True)
```

## Now You can feed any new data and model will predict the Species

```python
sepal_length=6.7
sepal_width=4.0
petal_length=6.1
petal_width=1.5

info = [[sepal_length,sepal_width,petal_length,petal_width]]
result = dtclassifier.predict(info)
print(result)
```
